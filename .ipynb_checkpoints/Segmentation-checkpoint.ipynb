```python
#imports 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Mall_Customers.csv")

data.rename(columns={'Annual Income (k$)':'Income', 'Spending Score (1-100)':'Score'}, inplace=True)

data.drop(['CustomerID'], axis = 1, inplace=True)

data.hist()
```

```python
data['Gender'].value_counts()
```

```python
x = data['Income']
y = data['Score']
plt.scatter(x, y)
plt.xlabel('Income')
plt.ylabel('Score')
plt.show()
```

```python
data.drop(['Gender'], axis = 1).corr()
```

```python
male_income = data.query('Gender == "Male"')['Income']
female_income = data.query('Gender == "Female"')['Income']
```

```python
plt.violinplot([male_income, female_income], showmeans=True, quantiles=[[0.25, 0.75], [0.25, 0.75]], vert = False)
```

```python
corr = data.drop(['Gender'], axis = 1).corr()
sns.heatmap(corr, annot = True)
```

```python
corr_male = data.query('Gender == "Male"').drop(['Gender'], axis = 1).corr()
sns.heatmap(corr_male, annot = True)
```

```python
corr_female = data.query('Gender == "Female"').drop(['Gender'], axis = 1).corr()
sns.heatmap(corr_female, annot = True)
```
